```python
#First....which variables are continuous, and which are discrete?
#Discrete and continuous variables are two types of quantitative variables: 
#Discrete variables represent counts (e.g. the number of objects in a collection). 
#Continuous variables represent measurable amounts (weight, speed, income)

In other words... 
#discrete variable is obtained by counting 
#continuous variable is obtained by measuring

#Discrete- 
#months_since_last_claim 
#months_since_policy_inception
#number_of_open_complaints
#number_of_policies

#Continuous
#customer_lifetime_value - 8041
#income- 5694
#monthly_premium_auto - 202
#total_claim_amount - 5105

#How can we differentiate using a function between the two?
```

```python
#range for discrete variables is between 6 and 100 
#write a function that says: 
#so if numericals.nunique() is >200, then it's discrete
#otherwise, it's continuous

#numericals =  np.where(data['INCOME'].isna(), data['predicted_income'], data['INCOME'])
#^ is it possible to do it with a np.where statement?

#this is Rebecca's solution:

def different(x):
    categorical = x.select_dtypes(include = object) #first splitting data into categorical and numerical
    numerical = x.select_dtypes(include = np.number)
    
    discrete=[]
    continuous= []
    
    for column in numerical:
        if x[column].nunique() <= 100:
            discrete.append(x[column])
        else:
            continuous.append(x[column])
    
    discrete_df = pd.DataFrame(discrete).T
    continuous_df = pd.DataFrame(continuous).T
    
    return discrete_df, continuous_df, categorical

different(customer_df)
#how do I apply this function? and does it need to be 10 or 100? 
```
